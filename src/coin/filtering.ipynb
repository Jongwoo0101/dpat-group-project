```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv("../src/data/filtered_BTC_data_2024.csv")  # 파일 이름을 알맞게 바꾸세요

# 5% 이상 상승한 날만 필터링
increased_df = df[(df['전일 종가 대비 변화량의 절댓값'] >= 5) & (df['전일 종가 대비 변화량'] > 0)]

# 결과 저장
increased_df.to_csv("btc_increased_over_5percent.csv", index=False, encoding='utf-8-sig')

print("5% 이상 상승한 날의 데이터를 'btc_increased_over_5percent.csv'로 저장했습니다.")
```
