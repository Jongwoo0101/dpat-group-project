```python
import pyupbit
import os
from dotenv import load_dotenv
```

```python
# 암호화폐 목록
print(pyupbit.get_tickers())
```

```python
# 최근 체결가격
print(pyupbit.get_current_price("KRW-BTC"))
```

```python
# 차트 데이터
df = pyupbit.get_ohlcv("KRW-BTC")
print(df.tail())
```

```python
# 매수 / 매도 호가
print(pyupbit.get_orderbook(ticker="KRW-BTC"))
```

```python
# 로그인
load_dotenv()

access = os.getenv('UPBIT_ACCEES_KEY')
secret = os.getenv('UPBIT_SECRET_KEY')
upbit = pyupbit.Upbit(access, secret)
```

```python
# 잔고 조회
print(upbit.get_balance("KRW-XRP"))
print(upbit.get_balance("KRW"))
print(upbit.get_balance())
```

```python
# 지정가 매도
print(upbit.sell_limit_order("KRW-XRP", 600, 20))
```

```python
# 지정가 매수
print(upbit.buy_limit_order("KRW-XRP", 613, 10))
```

```python
# 시장가 매수 / 매도 주문
print(upbit.buy_market_order("KRW-XRP", 10000))
print(upbit.sell_market_order("KRW-XRP", 30))
```

```python
# 미체결 주문 조회
upbit.get_order("KRW-BTC")
```

```python
# 완료된 주문 조회
print(upbit.get_order("KRW-BTC", state="done"))
```

```python
# 주문 취소
print(upbit.cancel_order('50e184b3-9b4f-4bb0-9c03-30318e3ff10a'))
```
